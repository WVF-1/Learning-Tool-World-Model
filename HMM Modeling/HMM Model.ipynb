```python
import json
import numpy as np

# ----------------------------
# Define HMM parameters
# ----------------------------

states = ["learning", "practicing", "mastered"]

pi = np.array([0.2, 0.6, 0.2])

A = np.array([
    [0.75, 0.25, 0.00],
    [0.00, 0.80, 0.20],
    [0.00, 0.00, 1.00]
])

base_mastery = {
    "learning": 0.35,
    "practicing": 0.60,
    "mastered": 0.90
}

difficulty_multiplier = {
    "easy": 1.0,
    "moderate": 0.9,
    "hard": 0.75
}

# ----------------------------
# Export to JSON for Base44
# ----------------------------

model_export = {
    "states": states,
    "initial_distribution": pi.tolist(),
    "transition_matrix": A.tolist(),
    "base_mastery": base_mastery,
    "difficulty_multiplier": difficulty_multiplier
}

with open("hmm_model_export.json", "w") as f:
    json.dump(model_export, f, indent=2)

print("Model exported successfully.")
```
